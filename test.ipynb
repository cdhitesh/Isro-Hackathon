```python
import joblib
import pandas as pd
import numpy as np
```

```python
model = joblib.load("random_forest_crop.pkl")
columns = joblib.load("columns.pkl")
label_encoder = joblib.load("label_encoder.pkl")
```

```python
def preprocess_input(user_input: dict):
    """
    user_input: dictionary of raw input values
    returns: processed dataframe ready for model
    """
    df = pd.DataFrame([user_input])

    # Align columns with training data
    df = df.reindex(columns=columns, fill_value=0)

    return df
scaler = joblib.load("scaler.pkl")
```

```python
def predict_crop(user_input: dict):
    processed_data = preprocess_input(user_input)

    # Predict class index
    pred_class = model.predict(processed_data)[0]

    # Predict probability
    pred_proba = model.predict_proba(processed_data)[0]

    # Decode label
    crop_name = label_encoder.inverse_transform([pred_class])[0]

    # Confidence score
    confidence = np.max(pred_proba)

    return {
        "predicted_crop": crop_name,
        "confidence": round(confidence * 100, 2)
    }
```

```python
def predict_with_threshold(user_input, threshold=0.6):
    result = predict_crop(user_input)

    if result["confidence"] / 100 >= threshold:
        return result
    else:
        return {
            "message": "Low confidence prediction",
            "confidence": result["confidence"]
        }
```

```python
def batch_predict(csv_path):
    df = pd.read_csv(csv_path)
    df = df.reindex(columns=columns, fill_value=0)

    preds = model.predict(df)
    crop_names = label_encoder.inverse_transform(preds)

    df["Predicted_Crop"] = crop_names
    return df
```
